```python
def dice_coef(y_true, y_pred):
    print(y_true.shape)
    y_true_f = tf.reshape(tf.dtypes.cast(y_true, tf.float32), [-1])
    print(np.unique(y_true_f))
    print(y_true_f.shape)
    y_pred_f = tf.reshape(tf.dtypes.cast(y_pred, tf.float32), [-1])
    print(np.unique(y_pred_f))
    intersection = tf.reduce_sum(y_true_f * y_pred_f)
    categories=np.dstack((y_pred_f+y_true_f>255,((y_pred_f == 255) & (y_true_f == 0)),((y_pred_f == 0) & (y_true_f == 0)),((y_pred_f == 0) & (y_true_f == 255))))
    print(categories.shape)
    print(categories[:,:,1])
    #tp,fp,tn,fn = (categories.sum((0,1)))/1000 
    tp,fp,tn,fn = (categories.sum((0,1)))
    print(tp,fp,tn,fn)
    print(tp+fp+tn+fn)
    return (2. * intersection + 1.) / (tf.reduce_sum(tf.square(y_true_f)) + tf.reduce_sum(tf.square(y_pred_f)) + 1.)
```
